```python
def pour_water(jug1_capacity, jug2_capacity, target_amount):
    # Initialize the two jugs to be empty
    jug1 = 0
    jug2 = 0

    # While the target amount is not achieved
    while jug1 != target_amount and jug2 != target_amount:
        # If jug1 is empty, fill it
        if jug1 == 0:
            jug1 = jug1_capacity
            print("Fill the", jug1_capacity, "gallon jug")
        # Pour water from jug1 to jug2 until either jug2 is full or jug1 is empty
        elif jug1 != 0 and jug2 != jug2_capacity:
            transfer_amount = min(jug1, jug2_capacity - jug2)
            jug1 -= transfer_amount
            jug2 += transfer_amount
            print("Pour", transfer_amount, "gallons from", jug1_capacity, "gallon jug to", jug2_capacity, "gallon jug")
        # If jug2 is full, empty it
        elif jug2 == jug2_capacity:
            jug2 = 0
            print("Empty the", jug2_capacity, "gallon jug")

    # Printing the final state of the jugs
    print("Final state: (", jug1, ",", jug2, ")")
```

```python
# Solving the specific problem with 4-gallon and 3-gallon jugs to get 2 gallons in the 4-gallon jug
print("Problem 1: Specific problem with 4-gallon and 3-gallon jugs to get 2 gallons in the 4-gallon jug")
pour_water(4, 3, 2)
```

```python
# Solving the generalized problem with customizable jug sizes and target amount
print("\nProblem 2: Generalized problem with customizable jug sizes and target amount")
# Example: Get 7 gallons in a 10-gallon jug
jug1_capacity = 10
jug2_capacity = 7
target_amount = 7
pour_water(jug1_capacity, jug2_capacity, target_amount)
```
